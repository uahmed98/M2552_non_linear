**Phase Portraits using Python**

The following code should allow you to plot phase portraits for the non-linear system

$$x' = f(x, y)$$
$$y' = g(x, y)$$

with f and g defined below. Just press SHIFT + ENTER on each of the highlighted cells or press the play button which should be somewhere above. Feel free to play around by changing up f and g! As written, we are working with the system

$$x' = x - y^2$$
$$y' = x + x^2 - 2y$$

Note that in Python, x**2 is the same as $x^2$.

```python
import numpy as np
from matplotlib import pyplot

def f(x, y):
    return x - y**2     ##You can modify this.
def g(x, y):
    return x + x**2 - 2*y   ##You can modify this. Note that x**2 is "x squared".

x_min = -1
x_max = 2
y_min = -1
y_max = 2

##Will plot the phase portrait for x between x_min and x_max and y between y_min and y_max.
```

```python
fine = 100
x = np.linspace(x_min, x_max, num = fine)
y = np.linspace(y_min, y_max, num = fine)


x_plot = np.array([f(x[j], y[i]) for i in range(fine) for j in range(fine)]).reshape(fine, fine)
y_plot = np.array([g(x[j], y[i]) for i in range(fine) for j in range(fine)]).reshape(fine, fine)

pyplot.streamplot(x, y, x_plot, y_plot)
```
